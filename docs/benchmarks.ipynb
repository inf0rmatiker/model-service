```python
import matplotlib.pyplot as plt

# ( name, year, PFS, PFS-day )
# Sorted by year
observed_results = [
    ("AlexNet", 2012.4, 500, 0.0058),
    ("Dropout", 2012.5, 184, 0.0021),
    ("Visualizing Conv Nets", 2013.85, 532, 0.0062),
    ("DQN", 2013.95, 2.3, 2.7e-5),
    ("VGG", 2014.7, 8524, 0.098),
    ("Seq2Seq", 2014.71, 7300, 0.085),
    ("DeepSpeech2", 2015.95, 26000, 0.3),
    ("ResNet", 2015.96, 10000, 0.117),
    ("Neural Machine Translation", 2016.7, 6.9e6, 79),
    ("Xception", 2016.75, 4.5e5, 5.0),
    ("Neural Architecture Search", 2016.8, 1.9e6, 22.0)
]

moores_law = [(2012, 500), (2013, 1000), (2014, 2000), (2015, 4000), (2016, 8000), (2017, 16000)]
estimate =   [(2012, 500)]
for i in range(1, len(moores_law)):
    estimate.append(
        (estimate[i-1][0]+1, 4 * estimate[i-1][1])
    )

print(estimate)

years    = [x[1] for x in observed_results]
pfs      = [x[2] for x in observed_results]
pfs_days = [x[3] for x in observed_results]

plt.figure(figsize=(15,7))
plt.plot(years, pfs, '.', label="AI Compute")
plt.plot([x[0] for x in moores_law], [y[1] for y in moores_law], label="Moore's Law Estimate")
plt.plot([x[0] for x in estimate], [y[1] for y in estimate], label="Line of Best Fit")
plt.yscale("log")
plt.title("Observed AI Improvements vs Moore's Law Trajectory")
plt.xlabel("Year")
plt.ylabel("PetaFLOPS")
plt.legend()
plt.savefig("./PFS_AI_vs_Moores.png", dpi=300)
```
